```python
def plot_function(word_freq):
    x = []
    y = []
    for i in word_freq:
        if len(i[0])== 2:
            x.append(i[0][0])
        else:
            x.append(i[0])
        y.append(i[1])

    x_id = np.arange(len(x))

    plt.plot(x_id,y, '-*')
    plt.show()
    plt.xticks(x_id,x, rotation = 65)
```

```python
def tweet_to_word_array(tweet):
    tweet = unquote(tweet)
    http_part = re.findall(r"htt\S*",tweet)
    if len(http_part) > 0:
        for k in range(len(http_part)):
            tweet = re.sub(re.escape(http_part[k]), '', tweet)
    url_part = re.findall(r"\/\/\S*",tweet)
    if len(url_part) > 0:
        for k in range(len(url_part)):
            tweet = re.sub(re.escape(url_part[k]), '', tweet)
    #print(tweet)
    words = nltk.word_tokenize(tweet)
    #words
    real_words = [w for w in words if w.lower() not in stopwords]
    #print(real_words)
    #parts_of_speech = nltk.pos_tag([j.lower() 
    #                                for j in real_words])
    parts_of_speech = nltk.pos_tag(real_words)
    #
    #print(parts_of_speech)
    lemma = WordNetLemmatizer()
    lemmatized = []
     
    for i in parts_of_speech:
        if i[1] in interested_pos:
            if i[1] in lemma_pos:
                lemmatized.append(lemma.lemmatize(i[0], pos = pos_dict[i[1]]))
            else:
                lemmatized.append((i[0]))
    #print(lemmatized)
    return lemmatized
    #return tweet
```
