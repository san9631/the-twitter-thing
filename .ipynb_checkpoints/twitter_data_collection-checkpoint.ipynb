```python
# q = '#LasVegasShooting' 
# q = '#LasVegas' 
# q = '#HarvestFestival'
# q = '#Vegas'
# q = '#prayforvegas'
# q = '#LasVegasShooter'
# q = '#StevenPaddock'
# q = '#LasVegasMassacre'
# q = 'las vegas gun control'
# q = 'las vegas shooting'
# q = '#Route91'
# q = 'VegasStrong'
q = 'HopeForVegas'
# Import unquote to prevent url encoding errors in next_results - you DON'T want this!
from urllib.parse import unquote #

count = 100 

search_results = twitter_api.search.tweets(q=q, count=count)

statuses = search_results['statuses']
print(len(statuses))
```

```python
print (json.dumps(statuses[1], indent=1))
```

```python
# Iterate through 1000 more batches of results by following the cursor 

for k in range(1, 1000):
    
    print("Length of statuses", len(statuses)) 
    if  k%100 == 0:
        time.sleep(15*60) # to follow rate limits and not to make twitter angry. 
    try:
        next_results = search_results['search_metadata']['next_results']
        # print('next_results')
    except KeyError as e: # No more results when next_results doesn't exist
        print('end of tweets')
        break 
        
        # Create a dictionary from next_results, which has the following form:
    # ?max_id=313519052523986943&q=NCAA&include_entities=1
    kwargs = dict([ kv.split('=') for kv in unquote(next_results[1:]).split("&") ])
    # print(kwargs)
    search_results = twitter_api.search.tweets(**kwargs)
    statuses += search_results['statuses']

# Show one sample search result by slicing the list...
print (json.dumps(statuses[0], indent=1))
```
